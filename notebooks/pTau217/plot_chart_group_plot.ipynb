```python
import kaleido #required

import plotly
plotly.__version__ #5.5.0

import pandas as pd 
import altair as alt 

import matplotlib.pyplot as plt
from matplotlib.legend_handler import HandlerTuple
from collections import defaultdict

import plotly.graph_objects as go
from plotly.subplots import make_subplots
import plotly.io as pio

from plot_chart_group_plot import concat_data_lmer, figure_plot, combine_plotly_charts

# Set a different default renderer
pio.renderers.default = 'colab'
```

```python
biomarker = "pTau217"
# folder_name = "outputs_e0.25_m55"
figs = []
for row_idx, efficacy in enumerate(["1.0", "0.50", "0.25"]):
# for row_idx, efficacy in enumerate(["1.0"]):
    # for col_idx, m in zip(["A", "B", "C"], ["55","33","11"]):
    for col_idx, m in zip(["A"], ["34"]):
        folder_name = f"./outputs_e{efficacy}_m{m}"
    # for m in ["44"]:
        # folder_name = f"./archived_simulations/outputs_e{efficacy}_m{m}"
        dfs = [concat_data_lmer(biomarker=biomarker, folder_name=folder_name, ratio=i*10) for i in range(1,11)]
        df = pd.concat(dfs)
        df["Experiment type"].unique()
        # the ratio between the between-subject variance and within-subject variance
        fig = figure_plot(df, plot_ci=True, title=f"({col_idx}{row_idx+1}) \u03B1={efficacy}, M={m}")
        figs.append(fig)
        # fig = figure_plot(df, plot_ci=True, title=f"\u03B1={efficacy}")
        # fig = figure_plot(df, title=f"pTau217 Power Estimation: \u03B1={efficacy}")
        # fig.update_layout(        
        #     title=dict(
        #         text='pTau217 Power Estimation'
        #         ),
        #         xaxis=dict(
        #             title='Within-subject and Between-subject Variance Ratio',
        #             tickmode = 'linear',
        #             tick0 = 0.0,
        #             dtick = 0.1,
        #         )
        # )
        # fig.write_image(f'figs/efficacy{efficacy}_m{m}_fig.png', scale=4, width=1000, height=500) 
```
